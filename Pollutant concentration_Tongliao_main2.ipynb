```python
# coding:utf-8
import xlrd
import xlsxwriter
import numpy as np
from numpy import *
```

```python
def read_excel(my_excel):
    mybook = xlrd.open_workbook(my_excel)
    mytable = mybook.sheets()[0]
    #表头list
    items=mytable.row_values(0)
    #数据
    date=[]
    values=[]
    for row in range(1,len(mytable.col_values(0))):
        row_temp=mytable.row_values(row)
        date.append(row_temp[0])
        values.append(row_temp[1:len(row_temp)])
    return  date,values
```

```python
def month_average(date,values):
    month_average=[]
    month=['01','02','03','04','05','06','07','08','09','10','11','12']
    row=0
    for j in month:
        month_average.append([])
        PM25_tongliao=[]
        PM25_jiaojian=[]
        PM25_yishu=[]
        PM10_tongliao=[]
        PM10_jiaojian=[]
        PM10_yishu=[]
        SO2_tongliao=[]
        SO2_jiaojian=[]
        SO2_yishu=[]
        CO_tongliao=[]
        CO_jiaojian=[]
        CO_yishu=[]
        NO2_tongliao=[]
        NO2_jiaojian=[]
        NO2_yishu=[]
        O3_tongliao=[]
        O3_jiaojian=[]
        O3_yishu=[]
        for i in range(len(date)):
            if date[i][4:6]==j:
                PM25_tongliao.append(values[i][0])
                PM25_jiaojian.append(values[i][1])
                PM25_yishu.append(values[i][2])
                PM10_tongliao.append(values[i][3])
                PM10_jiaojian.append(values[i][4])
                PM10_yishu.append(values[i][5])
                SO2_tongliao.append(values[i][6])
                SO2_jiaojian.append(values[i][7])
                SO2_yishu.append(values[i][8])
                CO_tongliao.append(values[i][9])
                CO_jiaojian.append(values[i][10])
                CO_yishu.append(values[i][11])
                NO2_tongliao.append(values[i][12])
                NO2_jiaojian.append(values[i][13])
                NO2_yishu.append(values[i][14])
                O3_tongliao.append(values[i][15])
                O3_jiaojian.append(values[i][16])
                O3_yishu.append(values[i][17])
        month_average[row].append(mean(PM25_tongliao))
        month_average[row].append(mean(PM25_jiaojian))
        month_average[row].append(mean(PM25_yishu))
        month_average[row].append(mean(PM10_tongliao))
        month_average[row].append(mean(PM10_jiaojian))
        month_average[row].append(mean(PM10_yishu))
        month_average[row].append(mean(SO2_tongliao))
        month_average[row].append(mean(SO2_jiaojian))
        month_average[row].append(mean(SO2_yishu))
        month_average[row].append(mean(CO_tongliao))
        month_average[row].append(mean(CO_jiaojian))
        month_average[row].append(mean(CO_yishu))
        month_average[row].append(mean(NO2_tongliao))
        month_average[row].append(mean(NO2_jiaojian))
        month_average[row].append(mean(NO2_yishu))
        month_average[row].append(mean(O3_tongliao))
        #month_average[row].append(mean(O3_jiaojian))
        month_average[row].append(mean(O3_yishu)) 
        row+=1    
    return month_average     
```

```python
def write_excel(month_average,year):
    headline=['date','PM25_tongliao','PM25_jiaojian','PM25_yishu',\
                 'PM10_tongliao','PM10_jiaojian','PM10_yishu',\
                 'SO2_tongliao','SO2_jiaojian','SO2_yishu',\
                 'CO_tongliao','CO_jiaojian','CO_yishu',\
                 'NO2_tongliao','NO2_jiaojian','NO2_yishu',\
                 'O3_tongliao','O3_jiaojian','O3_yishu']
    month=['01','02','03','04','05','06','07','08','09','10','11','12']
    workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/monthly_result.xlsx',{'nan_inf_to_errors': True})
    worksheet1=workbook.add_worksheet('2015')
    worksheet2=workbook.add_worksheet('2016')
    worksheet3=workbook.add_worksheet('2017')
    worksheet4=workbook.add_worksheet('2018')
    worksheet5=workbook.add_worksheet('2019')
    worksheet6=workbook.add_worksheet('2020')
    if year==2015:
        row=0
        col=0
        for item in headline:
            worksheet1.write(row,col,item)
            col+=1   
        for items in month_average:
            col=0
            worksheet1.write(row+1,col,'2015'+month[row])
            for item in items:
                worksheet1.write(row+1,col+1,item)
                col+=1
            row+=1
    elif year==2016:
        row=0
        col=0
        for item in headline:
            worksheet2.write(row,col,item)
            col+=1   
        for items in month_average:
            col=0
            worksheet2.write(row+1,col,'2016'+month[row])
            for item in items:
                worksheet2.write(row+1,col+1,item)
                col+=1
            row+=1
    elif year==2017:
        row=0
        col=0
        for item in headline:
            worksheet3.write(row,col,item)
            col+=1   
        for items in month_average:
            col=0
            worksheet3.write(row+1,col,'2017'+month[row])
            for item in items:
                worksheet3.write(row+1,col+1,item)
                col+=1
            row+=1
    elif year==2018:
        row=0
        col=0
        for item in headline:
            worksheet4.write(row,col,item)
            col+=1   
        for items in month_average:
            col=0
            worksheet4.write(row+1,col,'2018'+month[row])
            for item in items:
                worksheet4.write(row+1,col+1,item)
                col+=1
            row+=1
    elif year==2019:
        row=0
        col=0
        for item in headline:
            worksheet5.write(row,col,item)
            col+=1   
        for items in month_average:
            col=0
            worksheet5.write(row+1,col,'2019'+month[row])
            for item in items:
                worksheet5.write(row+1,col+1,item)
                col+=1
            row+=1 
    elif year==2020:
        row=0
        col=0
        for item in headline:
            worksheet6.write(row,col,item)
            col+=1   
        for items in month_average:
            col=0
            worksheet6.write(row+1,col,'2020'+month[row])
            for item in items:
                worksheet6.write(row+1,col+1,item)
                col+=1
            row+=1 
    workbook.close()
    
    
    
```

```python
if __name__=="__main__":
    my_excel=[]
    for i in range(2015,2020):
        my_excel.append('/Volumes/Jacqueline/全国环境监测站点数据/'+str(i)+'total_result.xlsx')
    date,values=read_excel(my_excel[4]) 
    month_average=month_average(date,values)
    write_excel(month_average,2019)
```
