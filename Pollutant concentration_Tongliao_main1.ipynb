```python
# coding:utf-8
import os
import csv
import xlsxwriter
from numpy import *
```

```python
#读取PM2.5,PM10,SO2,CO,NO2,O3每天csv数据
def read_csv(filename):
    workbook=csv.reader(open(file))
    tongliao=[]
    jiaojian=[]
    yishu=[]
    for i in workbook:
        tongliao.append(i[984])
        jiaojian.append(i[985])   
        yishu.append(i[986]) 
    return tongliao,jiaojian,yishu 
```

```python
#计算PM2.5,PM10,SO2,CO,NO2每天浓度均值（O3为最大8小时）
def target_values(tongliao,jiaojian,yishu):
    #判断数据表格有几个小时的数据
    hours=(len(tongliao)-1)/15
    #获取PM2.5,PM10,SO2,CO,NO2,O3在该天每个小时的数据
    PM25_tongliao=[]
    PM25_jiaojian=[]
    PM25_yishu=[]
    PM10_tongliao=[]
    PM10_jiaojian=[]
    PM10_yishu=[]
    SO2_tongliao=[]
    SO2_jiaojian=[]
    SO2_yishu=[]
    CO_tongliao=[]
    CO_jiaojian=[]
    CO_yishu=[]
    NO2_tongliao=[]
    NO2_jiaojian=[]
    NO2_yishu=[]
    O3_tongliao=[]
    O3_jiaojian=[]
    O3_yishu=[]
    for i in range(int(hours)):
        PM25_tongliao.append(tongliao[i*15+2])
        PM25_jiaojian.append(jiaojian[i*15+2])
        PM25_yishu.append(yishu[i*15+2])
        PM10_tongliao.append(tongliao[i*15+4])
        PM10_jiaojian.append(jiaojian[i*15+4])
        PM10_yishu.append(yishu[i*15+4])
        SO2_tongliao.append(tongliao[i*15+6])
        SO2_jiaojian.append(jiaojian[i*15+6])
        SO2_yishu.append(yishu[i*15+6])
        CO_tongliao.append(tongliao[i*15+14])
        CO_jiaojian.append(jiaojian[i*15+14])
        CO_yishu.append(yishu[i*15+14])
        NO2_tongliao.append(tongliao[i*15+8])
        NO2_jiaojian.append(jiaojian[i*15+8])
        NO2_yishu.append(yishu[i*15+8])
        O3_tongliao.append(tongliao[i*15+12])#O3均选取每日8小时滑动均值
        O3_jiaojian.append(jiaojian[i*15+12])
        O3_yishu.append(yishu[i*15+12])
    #滤掉空数据
    PM25_tongliao=list(filter(None,PM25_tongliao)) 
    PM25_jiaojian=list(filter(None,PM25_jiaojian))  
    PM25_yishu=list(filter(None,PM25_yishu))
    PM10_tongliao=list(filter(None,PM10_tongliao)) 
    PM10_jiaojian=list(filter(None,PM10_jiaojian))  
    PM10_yishu=list(filter(None,PM10_yishu))
    SO2_tongliao=list(filter(None,SO2_tongliao)) 
    SO2_jiaojian=list(filter(None,SO2_jiaojian))  
    SO2_yishu=list(filter(None,SO2_yishu))
    CO_tongliao=list(filter(None,CO_tongliao)) 
    CO_jiaojian=list(filter(None,CO_jiaojian))  
    CO_yishu=list(filter(None,CO_yishu))
    NO2_tongliao=list(filter(None,NO2_tongliao)) 
    NO2_jiaojian=list(filter(None,NO2_jiaojian))  
    NO2_yishu=list(filter(None,NO2_yishu))
    O3_tongliao=list(filter(None,O3_tongliao)) 
    O3_jiaojian=list(filter(None,O3_jiaojian))  
    O3_yishu=list(filter(None,O3_yishu))
    #计算PM2.5,PM10,SO2,CO,NO2,O3在该天的均值 
    PM25_tongliao=list(map(float,PM25_tongliao))
    PM25_jiaojian=list(map(float,PM25_jiaojian))
    PM25_yishu=list(map(float,PM25_yishu))
    PM10_tongliao=list(map(float,PM10_tongliao))
    PM10_jiaojian=list(map(float,PM10_jiaojian))
    PM10_yishu=list(map(float,PM10_yishu))
    SO2_tongliao=list(map(float,SO2_tongliao))
    SO2_jiaojian=list(map(float,SO2_jiaojian))
    SO2_yishu=list(map(float,SO2_yishu))
    CO_tongliao=list(map(float,CO_tongliao))
    CO_jiaojian=list(map(float,CO_jiaojian))
    CO_yishu=list(map(float,CO_yishu))
    NO2_tongliao=list(map(float,NO2_tongliao))
    NO2_jiaojian=list(map(float,NO2_jiaojian))
    NO2_yishu=list(map(float,NO2_yishu))
    O3_tongliao=list(map(float,O3_tongliao))
    O3_jiaojian=list(map(float,O3_jiaojian))
    O3_yishu=list(map(float,O3_yishu))
    '''
    if len(PM25_tongliao)==0:
        pm25_tongliao=' '
    else:
        pm25_tongliao=mean(PM25_tongliao)
    if len(PM25_jiaojian)==0:
        pm25_jiaojian=' '
    else:
        pm25_jiaojian=mean(PM25_jiaojian)
    if len(PM25_yishu)==0:
        pm25_yishu=' '
    else:
        pm25_yishu=mean(PM25_yishu)
    if len(PM10_tongliao)==0:
        pm10_tongliao=' '
    else:
        pm10_tongliao=mean(PM10_tongliao)
    if len(PM10_jiaojian)==0:
        pm10_jiaojian=' '
    else:
        pm10_jiaojian=mean(PM10_jiaojian)
    if len(PM10_yishu)==0:
        pm10_yishu=' '
    else:
        pm10_yishu=mean(PM10_yishu)
    if len(SO2_tongliao)==0:
        so2_tongliao=' '
    else:
        so2_tongliao=mean(SO2_tongliao)
    if len(SO2_jiaojian)==0:
        so2_jiaojian=' '
    else:
        so2_jiaojian=mean(SO2_jiaojian)
    if len(SO2_yishu)==0:
        so2_yishu=' '
    else:
        so2_yishu=mean(SO2_yishu)
    if len(CO_tongliao)==0:
        co_tongliao=' '
    else:
        co_tongliao=mean(CO_tongliao)
    if len(CO_jiaojian)==0:
        co_jiaojian=' '
    else:
        co_jiaojian=mean(CO_jiaojian)
    if len(CO_yishu)==0:
        co_yishu=' '
    else:
        co_yishu=mean(CO_yishu)
    if len(NO2_tongliao)==0:
        no2_tongliao=' '
    else:
        no2_tongliao=mean(NO2_tongliao)
    if len(NO2_jiaojian)==0:
        no2_jiaojian=' '
    else:
        no2_jiaojian=mean(NO2_jiaojian)
    if len(NO2_yishu)==0:
        no2_yishu=' '
    else:
        no2_yishu=mean(NO2_yishu)
    '''
    pm25_tongliao=mean(PM25_tongliao)
    pm25_jiaojian=mean(PM25_jiaojian)
    pm25_yishu=mean(PM25_yishu)
    pm10_tongliao=mean(PM10_tongliao)
    pm10_jiaojian=mean(PM10_jiaojian)
    pm10_yishu=mean(PM10_yishu)
    so2_tongliao=mean(SO2_tongliao)
    so2_jiaojian=mean(SO2_jiaojian)
    so2_yishu=mean(SO2_yishu)
    co_tongliao=mean(CO_tongliao)
    co_jiaojian=mean(CO_jiaojian)
    co_yishu=mean(CO_yishu)
    no2_tongliao=mean(NO2_tongliao)
    no2_jiaojian=mean(NO2_jiaojian)
    no2_yishu=mean(NO2_yishu)
    if len(O3_tongliao)==0:
        o3_tongliao=' '
    else:
        o3_tongliao=max(O3_tongliao)
    if len(O3_jiaojian)==0:
        o3_jiaojian=' '
    else:
        o3_jiaojian=max(O3_jiaojian)
    if len(O3_yishu)==0:
        o3_yishu=' '
    else:
        o3_yishu=max(O3_yishu)
    
    Target_Values=[pm25_tongliao,pm25_jiaojian,pm25_yishu,\
                    pm10_tongliao,pm10_jiaojian,pm10_yishu,\
                    so2_tongliao,so2_jiaojian,so2_yishu,\
                    co_tongliao,co_jiaojian,co_yishu,\
                    no2_tongliao,no2_jiaojian,no2_yishu,\
                    o3_tongliao,o3_jiaojian,o3_yishu]
    return Target_Values
```

```python
def write_xlsx(Target_Values,year,date): 
    headline=['date','PM25_tongliao','PM25_jiaojian','PM25_yishu',\
                 'PM10_tongliao','PM10_jiaojian','PM10_yishu',\
                 'SO2_tongliao','SO2_jiaojian','SO2_yishu',\
                 'CO_tongliao','CO_jiaojian','CO_yishu',\
                 'NO2_tongliao','NO2_jiaojian','NO2_yishu',\
                 'O3_tongliao','O3_jiaojian','O3_yishu']
    if year==2015:
        #创建结果文件
        workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/2015total_result.xlsx',{'nan_inf_to_errors': True})
        worksheet=workbook.add_worksheet('2015')
        #写入表头
        row=0
        col=0
        for item in (headline):
            worksheet.write(row, col, item)
            col += 1
        #追加PM2.5,PM10,SO2,CO,NO2,O3的每日均值
        row=1
        for item in (Target_Values):
            item.insert(0,date[row-1])
            for col in range(len(item)):
                worksheet.write(row, col, item[col])  
            row+=1
        print("2015-Write over!")
    elif year==2016:
        workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/2016total_result.xlsx',{'nan_inf_to_errors': True})
        worksheet=workbook.add_worksheet('2016')
        #写入表头
        row=0
        col=0
        for item in (headline):
            worksheet.write(row, col, item)
            col += 1
        #追加PM2.5,PM10,SO2,CO,NO2,O3的每日均值
        row=1
        for item in (Target_Values):
            item.insert(0,date[row-1])
            for col in range(len(item)):
                worksheet.write(row, col, item[col])  
            row+=1
        print("2016-Write over!")
    elif year==2017:
        workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/2017total_result.xlsx',{'nan_inf_to_errors': True})
        worksheet=workbook.add_worksheet('2017')
        #写入表头
        row=0
        col=0
        for item in (headline):
            worksheet.write(row, col, item)
            col += 1
        #追加PM2.5,PM10,SO2,CO,NO2,O3的每日均值
        row=1
        for item in (Target_Values):
            item.insert(0,date[row-1])
            for col in range(len(item)):
                worksheet.write(row, col, item[col])  
            row+=1
        print("2017-Write over!")
    elif year==2018:
        workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/2018total_result.xlsx',{'nan_inf_to_errors': True})
        worksheet=workbook.add_worksheet('2018')
        #写入表头
        row=0
        col=0
        for item in (headline):
            worksheet.write(row, col, item)
            col += 1
        #追加PM2.5,PM10,SO2,CO,NO2,O3的每日均值
        row=1
        for item in (Target_Values):
            item.insert(0,date[row-1])
            for col in range(len(item)):
                worksheet.write(row, col, item[col])  
            row+=1 
        print("2018-Write over!")
    elif year==2019:
        workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/2019total_result.xlsx',{'nan_inf_to_errors': True})
        worksheet=workbook.add_worksheet('2019')
        #写入表头
        row=0
        col=0
        for item in (headline):
            worksheet.write(row, col, item)
            col += 1
        #追加PM2.5,PM10,SO2,CO,NO2,O3的每日均值
        row=1
        for item in (Target_Values):
            item.insert(0,date[row-1])
            for col in range(len(item)):
                worksheet.write(row, col, item[col])  
            row+=1
        print("2019-Write over!")
    elif year==2020:
        workbook=xlsxwriter.Workbook('/Volumes/Jacqueline/全国环境监测站点数据/2020total_result.xlsx',{'nan_inf_to_errors': True})
        worksheet=workbook.add_worksheet('2020')
        #写入表头
        row=0
        col=0
        for item in (headline):
            worksheet.write(row, col, item)
            col += 1
        #追加PM2.5,PM10,SO2,CO,NO2,O3的每日均值
        row=1
        for item in (Target_Values):
            item.insert(0,date[row-1])
            for col in range(len(item)):
                worksheet.write(row, col, item[col])  
            row+=1
        print("2020-Write over!")
    workbook.close()
    print("Write over!")
```
